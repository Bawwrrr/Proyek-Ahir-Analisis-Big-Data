```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('/content/spotify_songs.csv')

data.info(10)
```

```python
print("\n--- Dataset Overview ---")
print("\n")
print("Shape of Dataset:", data.shape)
print("\n")
print("Column Names:", data.columns)
print("\n")
print("\n--- Sample Data ---")
print("\n")
display(data.head())
```

```python
print("\t\t--- Descriptive Statistics ---\n")
print(data.describe())
```

```python
# Artis dengan popularitas rata-rata tertinggi
artist_popularity = data.groupby('track_artist')['track_popularity'].mean().sort_values(ascending=False).head(10)

# Visualisasi
artist_popularity.plot(kind='bar', color='cyan')
plt.title('Top 10 Artists by Average Popularity')
plt.xlabel('Artist')
plt.ylabel('Average Popularity')
plt.xticks(rotation=45)
plt.show()
```

```python
popularity_genre = data.groupby('playlist_genre')['track_popularity'].mean().sort_values(ascending=False)

sns.barplot(x=popularity_genre.index, y=popularity_genre.values, palette="viridis")
plt.xticks(rotation=45)
plt.title('Average Popularity by Playlist Genre')
plt.xlabel('Genre')
plt.ylabel('Average Popularity')
plt.show()
```

<h2>Melihat dari hasil itu, melihat perbandingan genre dengan popularitas, Genre POP menjadi genre dengan popularitas tertinggi daripada genre yang lainnya

```python
# Korelasi antara fitur musikal
correlation_matrix = data[['danceability', 'energy', 'valence', 'tempo', 'loudness', 'track_popularity']].corr()

# Visualisasi korelasi
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix of Musical Features')
plt.show()
```

## Hubungan antar fitur musikal:

- energy dan loudness memiliki korelasi positif yang kuat (0.68). Artinya, lagu yang lebih energik cenderung lebih keras.
- valence dan danceability memiliki korelasi positif sedang (0.33). Ini berarti lagu yang lebih ceria (valence tinggi) cenderung lebih mudah untuk menari.
- tempo dan fitur lainnya memiliki korelasi lemah, artinya tempo tidak memiliki hubungan signifikan dengan danceability, energy, atau valence.

## Hubungan dengan track_popularity:

Tidak ada korelasi yang signifikan antara track_popularity (popularitas lagu) dengan fitur musikal. Nilai korelasinya semua mendekati 0, dengan hasil:
- Dengan danceability: 0.06 (lemah).
- Dengan energy: -0.11 (lemah negatif).
- Dengan valence: 0.03 (lemah).

Ini menunjukkan bahwa popularitas lagu tidak hanya ditentukan oleh fitur-fitur musikal, melainkan juga faktor lain seperti lirik, artis, atau promosi.


```python
# Konversi tanggal rilis ke format datetime
data['track_album_release_date'] = pd.to_datetime(data['track_album_release_date'], errors='coerce')

# Tren fitur musikal seiring waktu
data.set_index('track_album_release_date', inplace=True)
monthly_trends = data.resample('M')[['danceability', 'energy', 'valence']].mean()

monthly_trends.plot(figsize=(12, 6))
plt.title('Musical Features Over Time')
plt.xlabel('Time')
plt.ylabel('Average Value')
plt.legend(loc='upper left')
plt.show()
```

Melihat dari hasil yang ditampilkan untuk dianalisa, fitur dari ttahun 1959 hingga 2009 memiliki keragaman dalam bentuk fiturnya
- danceability memiliki nilai terendah menyentuh di angka 0.2 di tahun sebelum 1969 dan tertinggi di 0.8 ( melihat hasil seiringnya waktu berjalan lagu lagu di masa kini banyak yang menggunakan fitur dengan dance ability tinggi)
- lalu memiliki lalu energy memiliki nilai yang berada di angka 0.4 hingga hampir menyentuh angka 1 (semakin meningkatnya waktu energi dari lagu lagu mulai berisi energic)
- dalam valence juga ada beberapa nilai yang menyentuh di angka 0.2 dan tertinggi hingga hampir angka 1 (untuk emosi yang dibuat di dalam lagu, dalam seiringnya waktu banyak lagu yang dibuat netral sehingga tidak terlalu ceria dan tidak terlalu sedih)

```python
# Membandingkan lagu akustik dan tidak
acousticness_groups = data.groupby(pd.cut(data['acousticness'], bins=[0, 0.5, 1.0]))['track_popularity'].mean()

# Visualisasi
acousticness_groups.plot(kind='bar', color=['orange', 'green'])
plt.title('Popularity of Acoustic vs Non-Acoustic Songs')
plt.xlabel('Acousticness Range')
plt.ylabel('Average Popularity')
plt.show()
```

Melihat hasil dari visualisasi yang di tampilkan dari perbandingan rata-rata antara dua kelompok yaitu lagu akustik dengan diagram batang warna hijau dan tidak akustik bewarna orange.Dari grafik diatas terlihat bahwa :

*   Lagu dengan tidak akustik (warna oranye) memiliki popularitas rata-rata sekitar 40.
*   Lagu dengan akustik (warna hijau) memiliki popularitas rata-rata sedikit lebih tinggi, yaitu sekitar 45.

Dari hasil tersebut kita dapat melihat bahwa Lagu dengan  akustik cenderung sedikit lebih populer dibandingkan lagu dengan tidak akustik dalam dataset ini.

```python
# Distribusi durasi lagu
sns.histplot(data['duration_ms'] / 60000, bins=30, kde=True, color='blue')
plt.title('Distribution of Song Duration (in minutes)')
plt.xlabel('Duration (minutes)')
plt.ylabel('Frequency')
plt.show()

# Hubungan durasi dan popularitas
sns.scatterplot(x=data['duration_ms'] / 60000, y=data['track_popularity'])
plt.title('Duration vs Popularity')
plt.xlabel('Duration (minutes)')
plt.ylabel('Popularity')
plt.show()
```

Grafik histogram (Distribusi Durasi Lagu) menunjukkan bahwa:
  *   Distribusi frekuensi durasi lagu. Garis *kde=True* menambahkan kurva kepadatan kernel untuk memperlihatkan pola distribusi secara halus.
  *   Histogram ini membantu kita memahami seberapa sering durasi tertentu muncul di dataset.

Interpretasi:
*   Lagu dengan durasi tertentu cenderung mendominasi (misalnya, lagu dengan durasi antara 3-4 menit sering muncul di dataset musik umum).
*   Outlier mungkin muncul untuk lagu dengan durasi sangat pendek atau sangat panjang.

Grafik scatter (Hubungan Durasi dan Popularitas) menunjukkan bahwa:
*   Scatter plot memetakan durasi lagu (sumbu X) terhadap popularitas (sumbu Y).
*   Setiap titik pada plot mewakili sebuah lagu dalam dataset, dengan koordinat menunjukkan durasi dan tingkat popularitasnya.

Interpretasi:
*   Tidak terlihat hubungan linear yang jelas antara durasi lagu dan popularitas (berdasarkan sebaran titik yang cukup acak).
*   Lagu dengan durasi pendek (di bawah 4 menit) tampaknya lebih umum, dan popularitasnya bervariasi (dari 0 hingga 100).
*   Lagu dengan durasi lebih panjang (di atas 6 menit) jumlahnya lebih sedikit, tetapi popularitasnya tetap bervariasi.

Dari hasil visualisasi diatas kita bisa mengambil kesimpulan bahwa mayoritas lagu dalam dataset berdurasi sekitar 3-4 menit, mencerminkan standar umum di industri musik. Namun, analisis hubungan antara durasi dan popularitas menunjukkan tidak adanya korelasi signifikan. Popularitas lagu tampaknya lebih dipengaruhi oleh faktor lain, seperti genre atau artis, daripada durasinya.
